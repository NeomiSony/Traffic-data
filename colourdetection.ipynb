```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img=cv2.imread('road.jpg')
```

```
image_RGB=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
plt.figure(figsize=(20,8))
plt.imshow(image_RGB)
```

```
image_hsv = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2HSV)
```

```
lower_1 = np.array([0,150,50])
upper_1 = np.array([10,255,255])
```

```
sensitivity =15;
lower_2 = np.array([60 - sensitivity,100,100])
upper_2 = np.array([60 + sensitivity,255,255])
```

```
mask_1 = cv2.inRange(image_hsv, lower_1, upper_1)
```

```
mask_2 = cv2.inRange(image_hsv, lower_2, upper_2)
```

```
mask= mask_1 + mask_2
```

```
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```
res = cv2.bitwise_and(image_RGB, image_RGB, mask=mask)
```

```
plt.figure(figsize=(20,8))
plt.imshow(res)
```
